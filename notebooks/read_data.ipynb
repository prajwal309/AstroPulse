```python
from astropy.io import fits
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = '757137_TS.fits'
with fits.open(file) as data_psd:
    df_p = pd.DataFrame(data_psd[1].data)
    
plt.scatter(df_p.iloc[:,0].values, df_p.iloc[:,1].values, s=1, c='k')
plt.xlabel('Time (d)')
plt.ylabel('Flux (ppm)')
plt.show()
```
